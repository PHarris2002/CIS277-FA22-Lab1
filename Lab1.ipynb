# Lab 1 - Pandas Intro

```python
import pandas as pd

movies_df = pd.read_csv("movie.csv")
movies_df
```

## Questions


**0.** What is the average IMDB Score of all the movies?

```python
imdb_score_mean = movies_df.imdb_score.mean()

display(movies_df.loc[:,["movie_title","duration"]])

print(f"The average IMDB Score of all movies is {imdb_score_mean}")
```

**1.** How many movies have "Star Wars" in the movie_title field?

```python
star_wars_mask = movies_df['movie_title'].str.contains('Star Wars')
star_wars_movies = movies_df[star_wars_mask].loc[:, "movie_title"]
star_wars_movies_count = star_wars_movies.shape[0]

print(f"There are {star_wars_movies_count} movies with Star Wars in the movie_title field.")
display(star_wars_movies)
```

**2.** Which movie has the highest IMDB Score? What is that score?

```python
highest_imdb_score = movies_df.imdb_score == movies_df.imdb_score.max()
highest_imdb_score_filtered = movies_df[highest_imdb_score].loc[:, ["movie_title", "imdb_score"]]
highest_imdb_score_filtered
```

**3**. Which movie has the lowest IMDB Score? What is that score?


```python
lowest_imdb_score = movies_df.imdb_score == movies_df.imdb_score.min()
lowest_imdb_score_filtered = movies_df[lowest_imdb_score].loc[:, ["movie_title", "imdb_score"]]
lowest_imdb_score_filtered
```

**4.** For each director, how many movies have they directed (from this dataset)?

```python
print("Here are all the listed directors and the number of movies directed.")
display(movies_df.director_name.value_counts().sort_index())
```

**5.** What's the average duration of Action movies?

```python
action_mask = movies_df['genres'].str.contains('Action')
action_duration_average = movies_df[action_mask].duration.mean()

print(f"The average duration of action movies is {action_duration_average} minutes.")
```

**6.** Which PG-13 movie had the biggest budget?

```python
pg_13_mask =  movies_df.content_rating == "PG-13"
highest_budget_pg_13_movies = movies_df[pg_13_mask].loc[:, ["movie_title", "content_rating", "budget"]].sort_values("budget", ascending=False)

print("Here are PG-13 movies with the highest budgets. Princess Mononoke has the highest budget.")
highest_budget_pg_13_movies
```

**7.** How many movies were made in each year (title_year)?

```python
print("Here are the numbers of movies made for each year.")
display(movies_df.title_year.value_counts().sort_index())
```

**8.** How many movies had more than 10000 movie_facebook_likes?

```python
large_facebook_likes = movies_df[movies_df.movie_facebook_likes > 10000]
large_facebook_likes_filtered = large_facebook_likes.loc[:, ["movie_title", "movie_facebook_likes"]]
facebook_like_number = large_facebook_likes_filtered.shape[0]

print(f"There are {facebook_like_number} movies with over 10000 movie_facebook_likes")
display(large_facebook_likes_filtered)
```


**9.** How many movies have "Tom Hanks" (as actor1, actor2, or actor3)?

```python
tom_hanks_mask = (movies_df.actor_1_name == "Tom Hanks") | (movies_df.actor_2_name == "Tom Hanks") | (movies_df.actor_3_name == "Tom Hanks")
tom_hanks_movies_filtered = movies_df[tom_hanks_mask].loc[:, ["movie_title", "actor_1_name", "actor_2_name", "actor_3_name"]]

print(f"There are {tom_hanks_movies_filtered.shape[0]} movies that feature Tom Hanks as actor 1, 2, or 3.")
display(tom_hanks_movies_filtered)
```

**10.**  What are the movies and number of voted users (num_voted_users) for movies that have less than 2 faces in the poster (facenumber_in_poster)?

```python
print("Here are the movies that has a poster with less than two faces in it, as well as their number of voted users.")

less_than_two_face_movies = movies_df[movies_df.facenumber_in_poster < 2]
less_than_two_face_movies_filtered = less_than_two_face_movies.loc[:, ["movie_title", "num_voted_users", "facenumber_in_poster"]]
less_than_two_face_movies_filtered
```

**11.** What are the directors and number of director facebook likes for movies made beteween 2000 and 2004 (inclusive) sorted in descending order of director facebook likes?

```python
movies_before_2004 = movies_df[movies_df.title_year <= 2004]
movies_between_given_years = movies_before_2004[movies_df.title_year >= 2000]

print("Here are the directors and sorted director facebook likes for movies made between the years 2000 and 2004.")
display(movies_between_given_years.loc[:, ["movie_title", "title_year", "director_facebook_likes"]].sort_values("director_facebook_likes", ascending=False))
```

12. What are the missing values (NaN)?

```python
print("Here is data that holds missing values.")

missing_values_mask = movies_df.duration.isna()
movies_with_missing_values = movies_df[missing_values_mask]
movies_with_missing_values
```
